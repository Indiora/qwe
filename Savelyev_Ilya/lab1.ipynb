**1.Бинарный метод и метод множителей

```python
# Бинарный метод
def binary_method(x, n):
    # Двоичная запись без последней цифры
    bin_form = bin(n)[3::]
    number_of_operations = 0
    res = x
    for a in str(bin_form):
        # КУ
        if(a == '1'):
            # Умножить на себя
            res = res * res
            # Умножить на x
            res = res * x
            number_of_operations = number_of_operations + 2
        # К
        else:
            # Умножить на себя
            res = res * res
            number_of_operations = number_of_operations + 1

    return [res, number_of_operations]

# Метод множителей
def multiplier_method(x, n):
    # Разложение в произведение простых множителей
    F = factor(n)
    # Если число простое
    if F[0][0] == n:
        # Вызываем функцию с n - 1
        res = multiplier_method(x, n-1)
        # Результат домножаем на х
        res[0] = res[0] * x
        res[1] = res[1] + 1
        return res
    # Если непростое, разбиваем степень на произведение n = i * j
    i = binary_method(x, F[0][0])
    j = binary_method(i[0], n // F[0][0])
    return [j[0], i[1] + j[1]]

print("Бинарный метод при n = 65, x = 1")
print("Кол-во операций: " + str(binary_method(1, 65)[1]))
print("Метод множителей при n = 65, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 65)[1]) + '\n')

print("Бинарный метод при n = 63, x = 1")
print("Кол-во операций: " + str(binary_method(1, 63)[1]))
print("Метод множителей при n = 63, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 63)[1]) + '\n')

print("Бинарный метод при n = 67, x = 1")
print("Кол-во операций: " + str(binary_method(1, 67)[1]))
print("Метод множителей при n = 67, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 67)[1]) + '\n')

print("Бинарный метод при n = 99, x = 1")
print("Кол-во операций: " + str(binary_method(1, 99)[1]))
print("Метод множителей при n = 99, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 99)[1]) + '\n')

```
